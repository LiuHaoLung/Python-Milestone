### 3. 探索資料

#### 3-1. 每一間銀行最大的收盤價。

```python
# Multi-Level indexing func
# 抓取 Close 欄位，且是在 Stock Info 裡面

bank_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').max()
```

#### 3-2. 計算出每間銀行的 returns，並新建一個欄位叫 returns。
* The fomula is:

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

```python
# pct_change 可以計算增長率

returns = bank_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').pct_change()
```

```python
# 第一天沒有，因為沒有前一天的值可以進行比較

returns.head()
```

#### 3-3. 繪製 returns 欄位的 pairplot。

```python
# 第一天的值是 NaN，所以會顯示錯誤，因為要從 index 1 後開始繪製圖
# it can see the Citigroup actually had a huge stock crash 

sns.pairplot(returns[1:])
```

#### 3-4. 透過 returns 欄位，找到每間銀行在哪一天有最好及最差的 returns。

```python
# pandas func can call the day of the min

returns.idxmin()
```

```python
# pandas func can call the day of the max

returns.idxmax()
```

#### 3-5. 觀察 returns 的標準差，哪間銀行的股票風險最高。

```python
# if the std very large mean the stock is going up and down a lot

returns.std()
```

#### 3-6. 觀察在 2015 的 returns 的標準差，哪間銀行的股票風險最高。

```python
returns.loc['2015-01-01':'2015-12-31'].std()
```

#### 3-7. 繪製 Morgan Stanley 在 2015 年的 distplot。

```python
sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS'],color='green',bins=100)
```

#### 3-8. 繪製 CitiGroup 在 2008 年的 distplot。

```python
sns.distplot(returns.loc['2008-01-01':'2008-12-31']['C'],color = 'red', bins = 100)
```

#### 3-9. 繪製每間銀行收盤價的 lineplot。

```python
# Citigroup just crush at 2009

bank_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').plot()
```

```python
# plotly

bank_stocks.xs(key = 'Close', axis =1 ,level = 'Stock Info').iplot()
```

#### 3-10. 繪製 BAC 在 2008 年每個月收盤價平均值的 lineplot。

```python
plt.figure(figsize = (12,4))

# 先抓出在 bank_stocks 裡面的 Close
BAC = bank_stocks.xs(key = 'Close', axis =1 ,level = 'Stock Info')

# 從 Close 中抓出時間及 BAC
# 用 rolling can moving the average
# window 可以計算不同天數的平均，像是一個星期或是 60 天
BAC.loc['2008-01-01':'2009-01-01']['BAC'].rolling(window = 30).mean().plot(label = '30 Day Avg')

BAC.loc['2008-01-01':'2009-01-01']['BAC'].plot(label = 'BAC CLOSE')
plt.legend()
```

#### 3-11. 繪製每間銀行收盤價的 heatmap。

```python
sns.heatmap(bank_stocks.xs(key = 'Close', axis = 1,level = 'Stock Info').corr(),annot = True)
```

#### 3-12. 繪製每間銀行收盤價的 clustermap。

```python
sns.clustermap(bank_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').corr(),annot=True)
```

```python
# plotly

close_corr = bank_stocks.xs(key = 'Close', axis = 1,level = 'Stock Info').corr()
close_corr.iplot(kind = 'heatmap', colorscale = 'rdylbu')
```

#### 3-13. 繪製 BAC 在 2015-01-01 到 2016-01-01 的 candle plot。

```python
# 和上面不同，是要抓取整個 BAC，所以 key 就會是 BAC，Level 就會在 Bank Ticker

bank_stocks.xs(key = 'BAC', axis = 1 ,level = 'Bank Ticker').loc['2015-01-01':'2016-01-01'].iplot(kind='candle')
```

#### 3-14. 繪製 Morgan Stanley 在 2015 的 Simple Moving Averages。

```python
MS = bank_stocks.xs(key = 'Close', axis =1 ,level = 'Stock Info')

# 從 Close 中抓出時間及 MS

MS.loc['2015-01-01':'2016-01-01']['MS'].ta_plot(study = 'sma', periods = [13,21,55], title = 'Simple Moving Averages')
```

#### 3-15. 繪製 BAC 在 2015 的 Bollinger Band Plot。
